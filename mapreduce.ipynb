```python
import requests as req

url = 'https://www.sportsv.net/articles/89652'
url2 = 'https://www.sportsv.net/articles/92200'

r = req.get(url2)
print(r.text)
```

```python
import bs4
import re

html = r.text
root = bs4.BeautifulSoup(html, 'html.parser')

find = root.find_all('h1')
find += root.find_all('h2')
find += root.find_all('p')
data = ''
for i in find:
    #print(i.text)
    data += i.text
print(data)
data = re.sub('\W*','',data)
```

```python
#data = open("data.txt", 'r', encoding='utf-8').read()
#print(data)
```

```python
import jieba

#data = open("data.txt", 'r', encoding='utf-8').read()
dic = ['AnthonyDavis','AD']
jieba.load_userdict(dic)
words = jieba.lcut(data)
print(words)
```

```python
counts = {}
for word in words:
    if len(word) == 1:
        continue
    counts[word] = counts.get(word, 0) + 1
print(counts)
```

```python
items = list(counts.items())
items.sort(key = lambda s: s[1], reverse=True)
print(items)
```

```python
num = len(items)
word = [None] * num
count = [None] * num
words = []
counts = []
for i in range(num):
    word[i], count[i] = items[i]
    if count[i] >= 10:
        print("{0:<10}{1:<5}".format(word[i], count[i]))
        words.append(word[i])
        counts.append(count[i])
    else:
        break
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams["font.sans-serif"] = ["Microsoft YaHei"]
mpl.rcParams['axes.unicode_minus'] = False
x = words
y = counts
plt.barh(x, y)
plt.savefig('mapreduce.png')
plt.show()
```
